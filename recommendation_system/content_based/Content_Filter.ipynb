```python
def classificar_abdominais(abd):
    if abd < 15:
        return 'Fraco'
    elif abd <= 30:
        return 'Moderado'
    else:
        return 'Bom'

def classificar_imc(imc):
    if imc < 18.5:
        return 'Abaixo do peso'
    elif imc < 25:
        return 'Normal'
    elif imc < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidade'

av_fisica['nivel_fisico'] = av_fisica['abdominal'].apply(classificar_abdominais)
av_fisica['faixa_imc'] = av_fisica['imc'].apply(classificar_imc)

def atribuir_persona(row):
    if row['nivel_fisico'] == 'Bom' and row['faixa_imc'] == 'Normal':
        return 'Executor'
    elif row['nivel_fisico'] == 'Moderado' and row['faixa_imc'] == 'Sobrepeso':
        return 'Planejador'
    elif row['nivel_fisico'] == 'Fraco' and row['faixa_imc'] == 'Obesidade':
        return 'Analista'
    elif row['nivel_fisico'] == 'Fraco' and row['faixa_imc'] == 'Abaixo do peso':
        return 'Comunicador'
    elif row['nivel_fisico'] == 'Moderado' and row['faixa_imc'] == 'Normal':
        return 'Comunicador'
    else:
        return 'Executor'  # fallback

av_fisica['persona'] = av_fisica.apply(atribuir_persona, axis=1)
av_fisica = av_fisica.fillna(0)

print(av_fisica[['peso', 'altura', 'abdominal', 'imc', 'nivel_fisico', 'faixa_imc', 'persona', 'cliente_codigo']])
print(av_fisica['cliente_codigo', 'persona'])
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

av_fisica = av_fisica.replace([np.inf, -np.inf], np.nan).dropna(subset=['imc', 'nivel_fisico', 'persona'])
df_recomendações = df_recomendacoes.replace([np.inf, -np.inf], np.nan).dropna(subset=['intensidade', 'tipo', 'persona'])

# 1. Vetores dos clientes
encoder_clientes = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cliente_cat = encoder_clientes.fit_transform(av_fisica[['nivel_fisico', 'persona']])


scaler = StandardScaler()
imc_scaled = scaler.fit_transform(av_fisica[['imc']])

vetores_clientes = np.concatenate([imc_scaled, cliente_cat], axis=1)


# 2. Vetores dos exercícios
encoder_exercicios = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
exercicio_cat = encoder_exercicios.fit_transform(df_recomendações[['intensidade', 'tipo', 'persona']])

vetores_exercicios = exercicio_cat

dim_cliente = vetores_clientes.shape[1]
dim_exercicio = vetores_exercicios.shape[1]
max_dim = max(dim_cliente, dim_exercicio)

def pad_array(arr, target_dim):
    if arr.shape[1] < target_dim:
        padding = np.zeros((arr.shape[0], target_dim - arr.shape[1]))
        return np.hstack([arr, padding])
    return arr

vetores_clientes_pad = pad_array(vetores_clientes, max_dim)
vetores_exercicios_pad = pad_array(vetores_exercicios, max_dim)

similaridades = cosine_similarity(vetores_clientes_pad, vetores_exercicios_pad)

top_n = 3
recomendacoes = []

for idx, cliente_id in enumerate(av_fisica['cliente_codigo']):
    sim_scores = list(enumerate(similaridades[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[:top_n]
    for i, score in sim_scores:
        exercicio = df_recomendações.iloc[i]['exercicio']
        recomendacoes.append({
            'cliente_codigo': cliente_id,
            'exercicio': exercicio,
            'similaridade': score
        })

df_recomendacoes_final = pd.DataFrame(recomendacoes)
df_recomendacoes_final
```

```python

probabilidade_diversificacao = 0.3  # 30% chance de realizar um exercício fora da persona

# Função para simular quais exercícios foram realizados por cada cliente
def simular_exercicios_realizados(av_fisica, df_recomendacoes, probabilidade_diversificacao):
    realizacoes = []

    # Itera sobre cada cliente
    for cliente_id, persona_cliente, peso, altura, imc in zip(av_fisica['cliente_codigo'], av_fisica['persona'], av_fisica['peso'], av_fisica['altura'], av_fisica['imc']):

        # Itera sobre cada exercício
        for exercicio_id, nome_exercicio, persona_exercicio in zip(df_recomendacoes['exercicio'], df_recomendacoes['descricao'], df_recomendacoes['persona']):
            # Ajustar a probabilidade com base na persona
            probabilidade = 0.7 if persona_cliente == persona_exercicio else probabilidade_diversificacao

            # Ajustar a probabilidade com base em características físicas (IMC, peso, altura)
            if imc > 30:  # Exemplo: clientes com IMC maior que 30 podem ter maior probabilidade de realizar exercícios focados em emagrecimento
                probabilidade += 0.1
            elif peso < 50:  # Clientes com peso muito baixo podem preferir exercícios mais leves
                probabilidade -= 0.1

            # Garantir que a probabilidade fique entre 0 e 1
            probabilidade = np.clip(probabilidade, 0, 1)

            # Decidir se o cliente realizou o exercício com base na probabilidade
            if np.random.rand() < probabilidade:  # Se o valor gerado for menor que a probabilidade, o exercício foi realizado
                realizacoes.append({
                    'cliente_codigo': cliente_id,
                    'peso': peso,
                    'altura': altura,
                    'imc': imc,
                    'nome_do_exercicio': exercicio_id  # Garantir que estamos utilizando o nome correto do exercício
                })

    return pd.DataFrame(realizacoes)

# Gerando os exercícios realizados
df_exercicios_realizados = simular_exercicios_realizados(av_fisica, df_recomendacoes, probabilidade_diversificacao)

# Exibindo a simulação (quais exercícios foram realizados por cada cliente)
print(df_exercicios_realizados)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.preprocessing import MultiLabelBinarizer

df_clientes_exercicios = df_exercicios_realizados[['cliente_codigo', 'peso', 'altura', 'imc', 'nome_do_exercicio']]

# Criar um DataFrame com as informações dos exercícios para recomendação
df_exercicios = df_recomendacoes[['exercicio', 'descricao', 'persona']]  # Suponha que este seja o DataFrame com os exercícios disponíveis

# 1.1: Transformar os dados de exercícios em características numéricas (dummy encoding)
mlb = MultiLabelBinarizer()
X_exercicios = mlb.fit_transform(df_exercicios['descricao'].str.split())  # Transforma descrições em uma matriz binária

# 1.2: Criar o modelo KNN para recomendação de exercícios
knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(X_exercicios)

# Passo 2: Dividir os dados em treino e teste
train, test = train_test_split(df_clientes_exercicios, test_size=0.2, random_state=42)

# Passo 3: Função para recomendar exercícios para um cliente
def recomendar_exercicios(cliente):

    cliente_perfil = np.array([cliente['peso'], cliente['altura'], cliente['imc']]).reshape(1, -1)

    cliente_perfil_exercicios = np.concatenate([cliente_perfil, np.zeros((cliente_perfil.shape[0], X_exercicios.shape[1] - cliente_perfil.shape[1]))], axis=1)

    distances, indices = knn.kneighbors(cliente_perfil_exercicios)

    recomendacoes = df_exercicios.iloc[indices[0]]['exercicio'].values

    print(f"Recomendações para o cliente {cliente['cliente_codigo']}: {recomendacoes}")
    return recomendacoes

test_recomendacoes = test.apply(recomendar_exercicios, axis=1)
test['recomendacoes'] = test_recomendacoes


def calcular_metricas(test, recomendacoes_col):
    # Criar uma coluna de verdade para saber quais exercícios realmente foram realizados
    test['exercicios_realizados'] = test.apply(lambda x: df_exercicios_realizados[df_exercicios_realizados['cliente_codigo'] == x['cliente_codigo']]['nome_do_exercicio'].tolist(), axis=1)

    precisao_total = []
    recall_total = []
    f1_total = []

    for idx, row in test.iterrows():
        recomendados = set(row[recomendacoes_col])  # Exercícios recomendados
        realizados = set(row['exercicios_realizados'])  # Exercícios que o cliente realmente fez

        print(f"Cliente {row['cliente_codigo']}:")
        print(f"  Exercícios recomendados: {recomendados}")
        print(f"  Exercícios realizados: {realizados}")

        # Calcular as métricas
        precisao = len(recomendados.intersection(realizados)) / len(recomendados) if len(recomendados) > 0 else 0
        recall = len(recomendados.intersection(realizados)) / len(realizados) if len(realizados) > 0 else 0
        f1 = 2 * (precisao * recall) / (precisao + recall) if (precisao + recall) > 0 else 0

        print(f"  Precisão: {precisao:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}")

        precisao_total.append(precisao)
        recall_total.append(recall)
        f1_total.append(f1)

    # Calculando as médias
    precisao_media = np.mean(precisao_total)
    recall_media = np.mean(recall_total)
    f1_media = np.mean(f1_total)

    return precisao_media, recall_media, f1_media

precisao, recall, f1 = calcular_metricas(test, 'recomendacoes')

print(f'Precisão: {precisao:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```
